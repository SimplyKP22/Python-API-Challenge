### Store Part I results into DataFrame

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
city_df = city_df.dropna()
```

```python
city_df
```

```python
# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
# Store 'Lat' and 'Lng' into  locations 

locations = city_df[["Lat", "Lng"]].astype(float)

# Store Humidity as the weight.
weights = city_df["Humidity"].astype(float)

```

```python
# Plot Heatmap. Setting the basic outline for the map 
figure_layout = {
    'width': '900px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}

# Define the parameters for the map output to make the map easier to read. The zoom and center can be adjusted
# based on user preference
fig = gmaps.figure(layout=figure_layout,zoom_level=2,center=(15,25))

# Create the layer for the heatmap
heat_layer = gmaps.heatmap_layer(locations, weights=weights, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1.5)

# Add heat layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.

```python
# Create a new dataframe for the cities that would fit an ideal weather condition based on several conditions being met

ideal_city = city_df.loc[(city_df['Max Temp'] <= 85) & (city_df['Max Temp'] >= 70) & (city_df['Wind Speed'] <= 9)\
                        & (city_df['Cloudiness'] == 0)]

# Make sure that if the dataframe is modified, rows with Nan values are removed

ideal_city = ideal_city.dropna()


ideal_city
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = ideal_city

hotel_df['Hotel Name'] = ''

hotel_df
```

```python
# Create the parameters needed for the api search 
params = {
    "radius": 50000,
    "types": "hotel",
    "keyword": "hotel",
    "key": g_key,
}

# Loop through the hotel_df and run a lat/long search for each city that will be used for the api search 

for index, row in hotel_df.iterrows():
    # get lat, lng from df
    lat = row["Lat"]
    lng = row["Lng"]

    # change location each iteration
    params["location"] = f"{lat},{lng}"
    
    # Use the search term: "hotel" and our lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    hotel_response = requests.get(base_url, params=params).json()
    
    try:
        hotel_df.loc[index, 'Hotel Name'] = hotel_response['results'][0]['name']
        
    except (KeyError, IndexError):
        print("Missing field/result info...skip")
        
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations)

# Display figure
fig.add_layer(markers)
fig
```

```python
# Create an additional layer to represent the Hotel name, the city, and the country

# Use the inf_box_template to create the layer information
hotel_data_layer = gmaps.marker_layer(
    locations,info_box_content=[info_box_template.format(**row) for index, row in hotel_df.iterrows()]
)

fig.add_layer(hotel_data_layer)
# Display figure
fig
```
